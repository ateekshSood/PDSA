```python
#prime number from 1 to 100

def prime_checker(n : int) -> bool:
    
    if n==2:
        return True
    
    for i in range(2 , n):              #for very large output can iterate till sqrt of the number
        if (n % i ==0):
            return False    
    return True


def prime_print(n) :
    if not isinstance(n , int):
        raise TypeError("x must be an integer")
    
    print(f"  {[i for i in range(2 , n+1) if prime_checker(i)]}")
    
    
prime_print(100)    
    
```

```python
#computing gcd using eculedian algorithm

#gcd(m , n) :
'''
    To find the gcd we first select the greater number then divide it by the smaller number 
    and take its quotient and remainder and divide the quotient by its remainder until 
    we get remainder as 0. The final smaller number will be our gcd

'''


def gcd(m , n):
    
    if not isinstance(m , int) or not isinstance(n , int):
        raise TypeError("Please Enter Integers")
    
    
    try:
        smaller_number = min(m , n)
        larger_number = max(m , n)
        
        if(larger_number % smaller_number == 0):
            return print(f"{smaller_number}")
        
        else:
            gcd(smaller_number , larger_number%smaller_number)
        
    
    except :
        raise TypeError("must be integers")
    
gcd(8, 12)
```

```python
#binary search recursive

def binary_search(list_ , value , flag_iterative_recursive = False):
    if flag_iterative_recursive :
        
        left  , right = 0 , len(list_) - 1 
        while(left <= right):
            
            mid =  (left+right)//2
            
            if list_[mid] == value:
                return True
            
            elif list_[mid] > value:
                right = mid - 1
                
            else:
                left = mid + 1
                
        return False        
    
    else:
        mid = len(list_)//2
        
        if len(list_) == 0:
            return False
        
        if list_[mid] == value:
            return True
        
        elif list_[mid] > value:
            return binary_search(list_[:mid] , value)
        
        elif list_[mid] < value:
            return binary_search(list_[mid+1:] , value)
        
        
print(binary_search([1,2,3,4] , 4 , flag_iterative_recursive=True))
            
        
    #iterative better better space cuz no stack space so O(1) space 
    
    
```
